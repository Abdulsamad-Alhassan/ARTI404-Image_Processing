```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure

# 1. Load the moon image
img = data.moon()

# 2. Calculate the 3rd and 80th percentiles
# Percentile values must be between 0 and 100
p3, p80 = np.percentile(img, (3, 80))

# 3. Rescale the intensity
# in_range defines the "stretch" points based on your percentiles
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

# 4. Plot the results and the histogram
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Display the rescaled image
ax1.imshow(img_rescale, cmap='gray')
ax1.set_title('Rescaled Moon Image (3rd-80th Perc)')
ax1.axis('off')

# Plot the histogram of the rescaled image
ax2.hist(img_rescale.ravel(), bins=256, histtype='step', color='black')
ax2.set_title('Histogram of Rescaled Image')
ax2.set_xlabel('Intensity Value')
ax2.set_ylabel('Pixel Count')

plt.tight_layout()
plt.show()
```
