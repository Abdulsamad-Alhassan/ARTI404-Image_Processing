```python
# rotating an image
import numpy as np
import cv2 as cv

img = cv.imread('images/cameraman.png', 0) # 0 is the same as what I used before: cv.IMREAD_GRAYSCALE.
rows, cols = img.shape

rot = cv.getRotationMatrix2D((cols/2, rows/2),30, 0.6) #2 x 3 matrix ,rotate the image around it's center,rotate by 30 degree, scale it down to 60% of original size
another_image = cv.warpAffine(img, rot, (cols, rows)) #apply the matrix rot to the original image

cv.imshow('img', another_image)
cv.imwrite('rotation_out.jpg', another_image)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# performing sheer operation on the image
import numpy as np
import cv2 as cv

img = cv.imread('images/cameraman.png', 0)
rows, cols = img.shape

M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]]) #a matrix for sheering, for every pixel moved down on the y-axis, shift the x-coordinate by 0.5 pixels to the right.
# (int(cols*1.5), int(rows*1.5)) expands the output window by 50% in both width and height
# warpAffine function uses 3 x 3 matrices. It calculates the new position of every pixel based on the matrix M
sheared_img = cv.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))

cv.imshow('img', sheared_img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# displaying the negative of an image
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('images/cameraman.png', 0)# loading the image

negative_image = 256 - 1 - image # original equation: L = 256 - 1- r

# Display the original and negative images side by side
plt.subplot(1,2,1) # to display the original image on the left side
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('original image')
plt.axis('off')# Turn off axis labels

plt.subplot(122) # to display the negative image on the right side
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title('negative image')
plt.axis('off')# Turn off axis labels

plt.show()
```

```python
# displaying the log transformation of an image
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('images/cameraman.png', 0)# loading the image

# Log Transformation
c = 100
log_transformed = c * (np.log(image + 1)) # original equation: s = c.log(1+R)

log_transformed = np.uint8(log_transformed) # Convert to 8-bit unsigned integer format
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)  # Subplot for the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')  # Turn off axis labels

plt.subplot(1, 3, 2)   # Subplot for the log_transformed image
plt.imshow(cv2.cvtColor(log_transformed, cv2.COLOR_BGR2RGB))
plt.title("log_transformed")
plt.axis('off')# Turn off axis labels

inverse_log_transformed = c * (np.exp(log_transformed / c) - 1) # Inverse-Log Transformation

inverse_log_transformed = np.uint8(inverse_log_transformed) # Convert to 8-bit unsigned integer format

plt.subplot(1, 3, 3)  # Subplot for the inverse_log_transformed image
plt.imshow(cv2.cvtColor(inverse_log_transformed, cv2.COLOR_BGR2RGB))
plt.title("inverse_log_transformed")
plt.axis('off')# Turn off axis labels

plt.show()  
```

```python
# displaying the log transformation of an image
import cv2
import numpy as np

image = cv2.imread('images/cameraman.png', 0)# loading the image

# Apply gamma correction (e.g., gamma = 1.5)
gamma = 2
gamma2=4

adjusted_image = np.power(image / 255.0, gamma) * 255.0
adjusted_image = adjusted_image.astype(np.uint8)

adjusted_image2 = np.power(image / 255.0, gamma2) * 255.0
adjusted_image2 = adjusted_image2.astype(np.uint8)

plt.figure(figsize=(10, 5))  # Create a figure with a specified size
plt.subplot(1, 3, 1)  # Subplot for the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')# Turn off axis labels

plt.subplot(1, 3, 2)  # Subplot for the adjusted_image gamma = 2 
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.title("adjusted_image gamma = 2")
plt.axis('off')# Turn off axis labels

plt.subplot(1, 3, 3)  # Subplot for the Original Image with gamma = 4
plt.imshow(cv2.cvtColor(adjusted_image2, cv2.COLOR_BGR2RGB))
plt.title("Original Image gamma = 4")
plt.axis('off')# Turn off axis labels

plt.show()
```
