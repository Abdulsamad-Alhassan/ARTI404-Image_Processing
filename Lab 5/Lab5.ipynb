```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Task 1: 7x7 Box Filter
img = cv2.imread('moon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Apply 7x7 Box Filter
box_blur = cv2.blur(img, (7, 7))

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(box_blur), plt.title('7x7 Box Filter')
plt.show()
```

```python
#Task 2: Gaussian Filtering 
# Apply Gaussian blurs
gauss_5 = cv2.GaussianBlur(img, (5, 5), 0)
gauss_21 = cv2.GaussianBlur(img, (21, 21), 0)

#now display with the comparision between the two guassian filters
plt.figure(figsize=(15, 5))
plt.subplot(131), plt.imshow(img), plt.title('Original')
plt.subplot(132), plt.imshow(gauss_5), plt.title('5x5 Gaussian')
plt.subplot(133), plt.imshow(gauss_21), plt.title('21x21 Gaussian')
plt.show()
```

```python
#Task 3: Laplacian Sharpening

# 1-Read the image 
img = cv2.imread('Butterfly.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 2-Convert image to float32 
img_float = img.astype(np.float32) / 255.0

# 3-Use cv2.Laplacian() 
laplacian = cv2.Laplacian(img_float, cv2.CV_32F, ksize=3)

# 4-Sharpen the image
sharpened = np.clip(img_float - laplacian, 0, 1)

# 5-showing the results
plt.figure(figsize=(15, 5))
plt.subplot(131), plt.imshow(img_float), plt.title('Original')
plt.subplot(132), plt.imshow(np.clip(laplacian + 0.5, 0, 1)), plt.title('Laplacian Edge Map')
plt.subplot(133), plt.imshow(sharpened), plt.title('Sharpened Image')
plt.show()
```
